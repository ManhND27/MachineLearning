```julia
X=A[:,3:12]
y=A[:,2]
y=y.+1
```

```julia
function train(X,y)
    N,D= size(X)
    K=length(unique(y))
    μ=zeros(K,D)
    σ=zeros(K,D)
    θ=zeros(K)
    for k= 1: K
        θ[k]=sum(y.==k)/N
        X_k=X[y.==k,:]
        μ[k,:]=mean(X_k, dims=1)
        σ[k,:]=std(X_k, dims=1)
    end
    (θ,μ,σ)
end
```

```julia
(θ,μ,σ)= train(X,y)
```

```julia
# xây dựng hàm dự đoán
function predict(θ,μ,σ,Xnew)
    # dự đoán dựa vào công thức xác suất đầy đủ
    # p(y=k|x)=p(y=k)*p(x|y=k) log 2 vế để mất phép nhân chuyển về phép cộng đây là ước lượng hợp lý cực đại
    K= length(θ)
    t=zeros(K)
    for k= 1:K
        α=log(θ[k])
        tmp=(Xnew - μ[k,:])./σ[k,:] # chia từng giá trị của vector tử cho vector mẫu
        tmp1=tmp*tmp'
        β=sum(log.(σ[k,:])) - sum(tmp1/2)
        t[k]= α + β
    end
    argmax(t)
end
```

```julia
Xnew=X[400,:]
print(predict(θ,μ,σ,Xnew))
y[400]
```

```julia
# xây dựng hàm accuracy
function accuracy(y,X,θ,μ,σ)
    ŷ=zeros(length(y))
    for i=1:length(y)
        ŷ[i]=predict(θ,μ,σ,X[i,:])
    end
    acc = sum(y.== ŷ)/ length(y)
    return acc
end
```

```julia
acc=accuracy(y,X,θ,μ,σ)
```
