```julia

function loss(X, y,θ)
    predict= X*θ
    u= predict -y
    (u' * u)/length(y)
end
```

```julia
J0= loss(X,y,[0;0])
```

```julia
Jbest= loss(X,y, θ_best)
```

```julia
function bgd(X,y,α= 0.01, maxIterations=1000)::Array{Float64,2}
    N,D= size(X)
    θs= zeros(D, maxIterations)
    for i= 1: maxIterations-1
        ∇ = X'*(X*θs[:,i]-y)/N
        θs[:,i+1]=θs[:,i]- α*∇
    end
    θs
end
```

```julia
θs= bgd(X,y, 0.01, 10)
```

```julia
T=3000
θs= bgd(X,y,0.01,T)
```

```julia
ts= collect(1:T)
Js=map(t -> loss(X,y,θs[:,t]), ts)
Js[end]
```

```julia
plot(ts, Js, xlabel="Number of iterations", ylabel="Cost J")
```
