```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_decent(gradient, start, learn_rate, n_iter=500, tolerance=1e-06):
    vector = start
    buble = []
    for _ in range(n_iter):
        diff = -learn_rate*gradient(vector)
        if np.all(abs(diff) < tolerance):
            break
        vector += diff
        buble.append(vector)
    return vector,buble
```

```python
def gradient_decent(gradient, x, y, start=[0.5,0.5], learn_rate=0.01, n_iter = 100, tolerance=1e-6, dtype="float64"):
    if not callable(gradient):
        raise TypeError("gradient' must be callable")
#setup datatype
    dtype_ = np.dtype(dtype)
#converting data
    x = np.array(x, dtype=dtype_)
    y = np.array(y, dtype=dtype_)
    if x.shape[0] != y.shape[0]:
        raise ValueError("x and y lenght do not match")
    tolerance = np.array(tolerance, dtype=dtype_)
    if np.any(tolerance <= 0):
        raise ValueError("'tolerance' must be greate than zeros")
    #checking
    for _ in range(n_iter):
        diff = -learn_rate*np.array(gradient_decent(x, y, vector), dtype_)
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
    return vector if vector.shape else vector.item()
```
