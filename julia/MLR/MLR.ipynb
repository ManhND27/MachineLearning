```julia
using DelimitedFiles
using Plots
```

```julia
A = readdlm("./data/iris-train.txt")
```

```julia
X = A[:,2:end]
```

```julia
y = Int.(A[:,1])
```

```julia
K=length(unique(y))
```

```julia
X = [ones(length(y)) X]
```

```julia
/function loss(X,y,θ)
    α = sum(θ[y,:].*X,dims=2)
    s = sum(exp.(θ*X'),dims=1)
    β = sum(α) - sum(log.(s))
    J = -β/length(y)
    return J
end
```

```julia
N,D = size(X)
θ = zeros(K,D)
```

```julia
J = loss(X,y,θ)
```

```julia
function onehost(X,y)
    N,D = size(X)
    K=length(unique(y))
    δ = zeros(N,K) 
    for i = 1:K
        δ[:,i]=(y.==i)
    end
    return δ
end
```

```julia
function gradient(X,y,θ,δ)
    K=length(unique(y))
    u = exp.(θ*X')
    z = sum(u,dims=1)
    P = u ./ z
    ∇ = (P - δ')*X/length(y)
end
```

```julia
function train(X,y,θ,α,T)
    δ = onehost(X,y)
    θs = []
    push!(θs,θ)
    for i=1:T-1
        ∇ = gradient(X,y,θ,δ)
        θ = θ - α*∇
        push!(θs,θ)
    end
    return θs
end
```

```julia
θ = zeros(K,D)
θs = train(X,y,θ,0.01,1000)
```

```julia
ts = collect(1:1000)
js = map(t->loss(X,y,θs[t]),ts)
```

```julia
plot(ts,js)
```
