```julia
using DelimitedFiles
using Plots
```

```julia
A = readdlm("./data/admission.txt",',',Float64,'\n')
```

```julia
X = A[:,1:2]
```

```julia
y=A[:,3]
```

```julia
X = [ones(length(y)) X]
```

```julia
# build sigmoid function
function sigmoid(X,θ)
    α = X*θ
    g = 1 ./ (1 .+ exp.(-α))
    return g
end
```

```julia
# build  loss function
function loss(X,y,θ)
    α = sigmoid(X,θ)
    l = y'*log.(α) + (1 .- y)'*log.(1 .- α)
    return -l/length(y)
end
```

```julia
# build gradient function 
function gradient(X,y,θ)
    ∇ = (X'*(sigmoid(X,θ) - y))/length(y)
    return ∇
end
```

```julia
θ = [0;0;0]
∇ = gradient(X,y,θ)
```

```julia
function train(X,y,α=0.001,T=1000)
    N ,D = size(X)
    θs = zeros(D,T)
    for i = 1:T-1
        ∇ = gradient(X,y,θs[:,i])
        θs[:,i+1] = θs[:,i]-α*∇
    end
    return θs
end
```

```julia
N, D = size(X)
θ0 = zeros(D)
θs = train(X,y)
```

```julia
ts = collect(1:1000)
Js = map(t -> loss(X,y,θs[:,t]),ts)
```

```julia
plot(ts,Js)
```
